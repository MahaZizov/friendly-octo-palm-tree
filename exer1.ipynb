```python
housing_raw = pd.read_csv(filename2,
                          parse_dates = ['period_begin', 'period_end'],
                          dtype = {'total_homes_sold': 'Int32', 'region_name': 'category'})
```

```python
housing_raw.head()
```

```python
cols = pd.read_csv(filename2, nrows=0).columns
missing = set(['period_begin', 'period_end']) - set(cols)
if missing:
    print(f"Columns not found: {missing}")
else:
    print("All columns found.")
```

```python
housing_raw.info()
```

```python
ca_housing = housing_raw.loc[
    (housing_raw['region_name'].str.contains('CA'))
].assign(region_name = housing_raw['region_name'].str[0 : -11])
```

```python
ca_housing.head()
```

```python
ca_housing_pivot = ca_housing.pivot_table(index = 'period_begin',
                                          columns = 'region_name',
                                          values = 'median_active_list_price').assign(CA_average = lambda x: x.mean(axis=1))
```

```python
ca_housing_pivot.head()
```

```python
ca_housing_markets = ca_housing_pivot.loc[:, ['San Francisco',
                                              'Los Angeles',
                                              'San Diego',
                                              'CA_average']]
ca_housing_markets.head()
```

```python
colors=["orange", "gray", "gray", "gray"]
ls = ["-", "--", ":", "-."]

fig, ax = plt.subplots(figsize=(10, 6))

for i, column in enumerate(ca_housing_markets.columns):
    ax.plot(
        ca_housing_markets[column] / 1000000,
            color=colors[i],
            ls=ls[i]
    )

#ax.plot(ca_housing_markets)

'''ax.plot(ca_housing_markets.index,
    ca_housing_markets['San Francisco'] / 1000000,
        label = 'San Francisco')


ax.plot(ca_housing_markets.index,
        ca_housing_markets['Los Angeles'] / 1000000,
        label = 'Los Angeles')'''

#ax.plot(ca_housing_markets['CA_average'])


fig.suptitle('Sth important', y=.95)
ax.set_title('CA housing prices 2017-2022')
ax.set_xlabel('Year')
ax.set_ylabel('Median Home Price')
ax.axvline(18341, c='gray', linestyle='--', label = 'Start of Covid')
#ax.text(17850, 1, "Start of Covid ===>")
ax.annotate("Start of Covid",
            xy=(18300,1),
            xytext=(17500, 1),
            arrowprops=dict(facecolor="black",
                            width=1,
                            headwidth=8,
                            connectionstyle="angle3, angleA=270, angleB=0"),
            verticalalignment="center")

ax.legend(['SF', 'LA', 'Start of Covid'], loc = 'best', frameon=False)
ax.set_ylim(0, 1.6)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
